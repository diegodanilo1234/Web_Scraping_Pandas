```python
# Cremos el root
root = "https://www.football-data.co.uk/mmz4281/"
```

### Multiples ligas

- https://www.football-data.co.uk/mmz4281/2122/E0.csv
- https://www.football-data.co.uk/mmz4281/2122/E1.csv
- https://www.football-data.co.uk/mmz4281/2122/E2.csv
- https://www.football-data.co.uk/mmz4281/2122/E3.csv
- https://www.football-data.co.uk/mmz4281/2122/EC.csv

```python
# Cremoa la variable ligas
leagues = ["E0", "E1", "E2", "E3", "EC"]
frames = []

for league in leagues:
  df = pd.read_csv(root + "2122" + "/" + league + ".csv", encoding="unicode_escape")
  frames.append(df)
```

```python
# Tamaño de frames
len(frames)
```

```python
# Mostrar el 1er elemento
frames[0]
```

# 2.1 Multiples temporadas

```python
leagues = ['E0','E2', 'E3']

frames = []

# hacer loop a un rango de temporadas e insertar una nueva columna
for league in leagues:
    for season in range(15,21): # rango de temporadas
        df = pd.read_csv(root + str(season) + str(season+1) + "/" + league + ".csv")
#         df['season'] = season
        df.insert(1, 'season', season)
        frames.append(df)
```

```python
# total de frames (6 temporadas x 3 ligas)
len(frames)
```

```python
# mostrar 1er y ultimo elemento (columna season fue creada)
frames[17]
```

# Organizar Data en un Diccionario

```python
# crear un diccionario 
dict_countries = {
              'Spanish La Liga':'SP1', 'Spanish Segunda Division':'SP2',
              'German Bundesliga':'D1', 
              'English Premier League':'E0', 'English League 1':'E2', 'English League 2':'E3',
             }
```

```python
# obtener un elemento del diccionario
dict_countries['German Bundesliga']
```

```python
dict_historical_data = {} 

# hacer loop sobre llaves de diccionario
for league in dict_countries:
    frames = [] # un dataframe por liga
    for season in range(15, 21):
        df = pd.read_csv(root + str(season) + str(season+1) + "/" + dict_countries[league]+ ".csv")
        df.insert(1, 'season', season)
        frames.append(df)
    df_frames = pd.concat(frames) # almacenar todas las temporadas de la liga en un solo dataframe
    dict_historical_data[league] = df_frames # asignar dataframe a un nuevo diccionario
```

```python
# mostrar dataframe 
dict_historical_data['Spanish La Liga']
```
